```python
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
meas = np.load('file1.npy')
```

```python
meas_info = open('file1.json')
json.load(meas_info)
```

```python
# fft
meas_f = np.fft.rfft(meas)
```

```python
# add coherence
FRF_estimate = 'H1' # or H2
channels = np.arange(meas_f.shape[1])
force_chn = 0
resp_chn = channels[np.where(channels != force_chn)]    
```

```python
def get_FRF(FRF_estimate='H1'):
    if FRF_estimate == 'H1':
        S_FX = np.zeros((len(resp_chn), meas_f.shape[-1]), dtype=complex)
        S_FF = np.zeros((len(resp_chn), meas_f.shape[-1]), dtype=complex)
    elif FRF_estimate == 'H2':
        S_XX = np.zeros((len(resp_chn), meas_f.shape[-1]), dtype=complex)
        S_XF = np.zeros((len(resp_chn), meas_f.shape[-1]), dtype=complex)
    else:
        print('Invalid estimate entry. Possible estimates: H1, H2')
        return
    for i in range(meas_f.shape[0]):
        F = meas_f[i, force_chn, :]
        X = meas_f[i, resp_chn, :]
        if FRF_estimate == 'H1':
            S_FX += np.conj(F)*X
            S_FF += np.conj(F)*F
        elif FRF_estimate == 'H2':
            S_XX += np.conj(X)*X
            S_XF += np.conj(X)*F
    if FRF_estimate == 'H1':
        H = S_FX/S_FF
    elif FRF_estimate == 'H2':
        H = S_XX/S_XF
    return H
```

```python
def get_H1_H2_coh():
    return get_FRF('H1')/get_FRF('H2')
```

```python
plt.plot(abs(get_H1_H2_coh().T))
plt.xlim(0, 2000)
```
